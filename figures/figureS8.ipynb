# figureS8

Run this notebook to generate `figureS8.svg` from `../results_processed/`.

```python
from pathlib import Path
import sys

FIGURE_ID = "figureS8"

def find_figures_dir():
    cwd = Path.cwd().resolve()
    candidates = [cwd, *cwd.parents]
    for path in candidates:
        if (path / "_plotting.py").exists():
            return path
        if (path / "figures" / "_plotting.py").exists():
            return path / "figures"
    raise FileNotFoundError("Could not find figures/_plotting.py")

figures_dir = find_figures_dir()
if str(figures_dir) not in sys.path:
    sys.path.insert(0, str(figures_dir))

from _plotting import render_figure

_ = render_figure(FIGURE_ID)
```
